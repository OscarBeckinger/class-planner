```python
#gets units of class when provided with abreviation ex: getUnits('ENGL 122') returns 3
def getUnits(classAbv):
    cl = df.loc[classAbv]  #getting row of class info (series) from dataframe
    units = float(cl.Units) #gets Units from row and casts to float
    return units
```

```python
#gets a class name as input and returns a list of one or more prereqs (list of strs) DOES NOT INCLUDE COREQ MARKING
def getPrereqs(classAbv):
    cl = df.loc[classAbv]
    prereqs = cl.Prereq.split('-')  #not worrying abt coreqs rn so removing ?s and splitting if mulitple prereqs
    return prereqs

#------------------------------Example                            
test = getPrereqs('PHYS 231')
test
```

```python
def getCoreqs(classAbv):
    cl = df.loc[classAbv]
    coreqs = cl.Coreq.split('-')  #not worrying abt coreqs rn so removing ?s and splitting if mulitple prereqs
    return coreqs

#------------------------------Example                            
test = getCoreqs('PHYS 125')
test
```

```python
def hasPrereq(classAbv):
    ph = getPrereqs(classAbv)
    if ph[0] == 'N':
        return False
    return True

#------------------------------Example
n = hasPrereq('ENGL 126')
n
```

```python
def hasCoreq(classAbv):
    ph = getCoreqs(classAbv)
    if ph[0] == 'N':
        return False
    return True

#--------------------------Example
n = hasCoreq('PHYS 125')
n
```

```python
#colleges is list of dictionaries (not strings) for Ex: [berkeley, ucla], majors is a list strings, each being a major name
#correspoding to the college in the same position of the college array as the major. Also removes any repeat courses.
#colleges = [c1, c2]  majors = ['m1', 'm2'] (input lists must be same leng). Returns a NumPy array.
def getStudentRequiredCourses(colleges, majors):
    sRq = []
    for i, colls in enumerate(colleges):
        sRq += colls[majors[i]]['reqCourses']
    return np.unique(sRq)
```

```python
#same as above just for recommended courses
def getStudentRecommendedCourses(colleges, majors):
    sRc = []
    for i, colls in enumerate(colleges):
        sRc += colls[majors[i]]['recCourses']
    return np.unique(sRc)
```

```python
#input is list of courses, returns new list of courses (maybe be unaltered though)
#function adds prereqs that arent in required courses list that need to be
def addPrereqs(courses):
    newCourses = courses
    for course in courses:  
        ph = getPrereqs(course)   #gets prereq list of a course
        if ph[0] == 'N':          #if list contains 'N' it means no prereqs
            continue
        else:                    #else go through list of prereqs, check if they are in the list of all courses, if not append to prereq to the list of all courses
            for prereq in ph:
                if prereq not in newCourses:
                    newCourses = np.append(newCourses, prereq)    
    return np.sort(newCourses)            #not sure if sorting here is useful yet (maybe take out later?)

#------------------------------------------- Example (of this and previous two functions)
cou = [ucla]
maj = ['Business Economics']
sC = getStudentRequiredCourses(cou, maj)
print(sC)
nsC = addPrereqs(sC)
print('new')
print(nsC)
```

```python
def addCoreqs(courses):
    newCourses = courses
    for course in courses:  
        ph = getCoreqs(course)   #gets coreq list of a course
        if ph[0] == 'N':          #if list contains 'N' it means no coreqs
            continue
        else:                    #else go through list of coreqs, check if they are in the list of all courses, if not append to the list of all courses
            for coreq in ph:
                if coreq not in newCourses:
                    newCourses = np.append(newCourses, coreq)    
    return np.sort(newCourses)            #sorting to make sure order is kept

#---------------------------------------------------- example
cul = [berkeley]
mij = ['Architecture']
sp = getStudentRequiredCourses(cul, mij)
print(sp)
nsp = addCoreqs(sp)
print('new')
print(nsp)
```

```python
#input is list/arr of strs each being a course name, returns a dictionary of courses sorted by their deparment ex: {'BUS': ['BUS 109', 'BUS 110'], etc}
def splitCourses(courses):
    course_dict = {}
    for course in courses:
        abbreviation = course.split()[0]            #split str into list to get abbreviation of course (MATH 240 becomes MATH) to be the keys for course dictionary
        if abbreviation in course_dict:             #if the abbreviation (key) is already there you can add the class to the arr corresponding to said key        
            course_dict[abbreviation] = np.append(course_dict[abbreviation], course)
        else:                                       #if not found create a key pair value with abbreviation and a new arr with the course starting it
            course_dict[abbreviation] = np.array([course])

    return course_dict

#---------------------------------example (sp defined in example above)
tt = splitCourses(sp)
tt
```

```python
#input: course/department dictionary, and checks length of values in it to see if there are any classes left
def checkLengOfVals(course_dict):
    ph = list(course_dict.values())    #get a list of all the values (list of numpy arrays)
    #print('\n\nOUTSIDE FUNC PRINTING TEST VALS')
    #print(len(ph))
    #print(ph)
    #print('\n\n')
    start = []
    for i in ph:                      #use for loop to extarct array vals and create one big arr of strings instead of list of arrays
        start = np.append(start, i)
    if len(start) > 0:              #check if there are any values in the array return accordingly
        return True
    else:
        return False
    
```

```python
def removeCourseFromDict(course_dict, course):
    abv = course.split()[0]
    clist = course_dict[abv]
    for i, c in enumerate(clist):
        if c == course:
            index = i
    course_dict[abv] = np.delete(course_dict[abv], index)
    return course_dict
#-------------------------------------------example (tt defined two cells above)
ttv2 = removeCourseFromDict(tt, 'ENGL 122')
ttv2
```

```python
def canClassBeAdded(classAbv, semesters, i):
    if not hasPrereq(classAbv):          #if theres no prereqs true will be returned right away
        return True
    if i == 0:                          #if you are in the 1st sem and there are prereqs you can't add the class (false returned)
        return False
    
    pres = getPrereqs(classAbv)
    if len(pres) == 2:                  #2 is max leng of prereqs. So each class only has 1 or 2 prereqs, but can also have coreqs
        if i == 1:
            if pres[0] in semesters[i-1].astype(str) and pres[1] in semesters[i-1].astype(str):
                return True
            else:
                return False
        elif i == 2:
            if (pres[0] in semesters[i-2].astype(str) or pres[0] in semesters[i-1].astype(str)) and (pres[1] in semesters[i-2].astype(str) or pres[1] in semesters[i-1].astype(str)):
                return True
            else:
                return False
        elif i == 3:
            if (pres[0] in semesters[i-3].astype(str) or pres[0] in semesters[i-2].astype(str) or pres[0] in semesters[i-1].astype(str)) and (pres[1] in semesters[i-3].astype(str) or pres[1] in semesters[i-2].astype(str) or pres[1] in semesters[i-1].astype(str)):
                return True
            else:
                return False
    else:                                 #the else case is for when there is only one prereq
        if i == 1:
            if pres[0] in semesters[i-1].astype(str):
                return True
            else:
                return False
        elif i == 2:
            if pres[0] in semesters[i-2].astype(str) or pres[0] in semesters[i-1].astype(str):
                return True
            else:
                return False
        elif i == 3:
            if pres[0] in semesters[i-3].astype(str) or pres[0] in semesters[i-2].astype(str) or pres[0] in semesters[i-1].astype(str):
                return True
            else:
                return False
        
    return True

#---------------------------------------------------------example

testsems = [np.array(['PSYCH 101', 'ENGL 122', 'MATH 142']), np.array(['PHYS 130','PSYCH 214', 'ARCHI 121']), np.array(['ENGL 126', 'ARCHI 130']), np.array([])]
i = 3
cb = 'PSYCH 215'
print(getPrereqs(cb))
testn = canClassBeAdded(cb, testsems, i)
testn
```

```python
#input is arr of required courses, output is a 4 semester plan.
def createPlan(reqCourses):
    sem1, sem2, sem3, sem4 = np.array([]), np.array([]), np.array([]), np.array([])
    semesters = [sem1, sem2, sem3, sem4]            #list of numpy array to hold course for each semester
    reqCourses = addPrereqs(reqCourses)             #adding any needed prereqs to the required courses list
    reqCourses = addCoreqs(reqCourses)              ##adding any needed coreqs to the required courses list
    amtOfCourses = len(reqCourses)                  #getting int number of courses ex: 14
    avgSemCourses = int(math.ceil(amtOfCourses/4))   #dividing number by 4 and rounding up to whole number to get approx number of courses per semester
    print('COURSES PER SEM: ', avgSemCourses)
    #allCourses = set(reqCourses)                    #creating a set to store all classes so i can remove dynamically as they are added to plan and also another check to make sure there are no repeats (sets cant have duplicates)
    coursesSplit = splitCourses(reqCourses)         #creating a dictionary of arrs splitting up courses by department 
    departments = list(coursesSplit.keys())         #getting a list of the keys from the dictionary (list of deparments ex ['BUS', 'COMSC', etc])
    i, courseCount = 0, 0
    #print(coursesSplit)
    lastClass = ''
    classAddedInNextSem = False
    while(checkLengOfVals(coursesSplit)):             #continue while there are more classes left in course deparment dictionary
        for d in departments:
            if len(coursesSplit[d]) != 0:                      #checking if list in courses dictionary has values in it
                
                #individual prereq check
                if i == 0:                                 #semester one Prereq check
                    prereqAdded = False
                    if hasPrereq(coursesSplit[d][0]):      #check if the individual course has prereqs
                        ph = getPrereqs(coursesSplit[d][0]) #if so we retrieve them
                        for pres in ph:                     #go through each prereq
                            if pres not in semesters[i]:    #make sure prereq is not already in the course plan
                                semesters[i] = np.append(semesters[i], pres) #if not then append it to the course plan
                                coursesSplit = removeCourseFromDict(coursesSplit, pres) #after the courses is added we must remove it from course dictionary
                                prereqAdded = True                           #tell the program a prereq was aded
                                courseCount +=1                             #increase total course count by 1
                                if courseCount % avgSemCourses == 0:       #check if its need to go to next semester
                                    i+=1
                    
                    if not prereqAdded:                                                #if not prereq was added do the followiing
                        if hasCoreq(coursesSplit[d][0]):                  #!          #check for coreqs
                            ch = getCoreqs(coursesSplit[d][0])            #!          #get coreqs
                            for cos in ch:                                         
                                if cos not in semesters[i]:                         #if  coreq not in plan add it
                                    semesters[i] = np.append(semesters[i], cos)
                                    courseCount +=1                                 #inc total course count
                                    coursesSplit = removeCourseFromDict(coursesSplit, cos)
                                    if courseCount % avgSemCourses == 0:        
                                        i+=1     
                        if coursesSplit[d][0] == lastClass:                                   #adding this to fix a infinite loop glitch that happens if only one department is left in the dictionary and the class trying to be added has a prereq in the same semester it is in so you must move to the next semester
                            semesters[i+1] = np.append(semesters[i+1], coursesSplit[d][0])
                            classAddedInNextSem = True
                            coursesSplit[d] = np.delete(coursesSplit[d], 0)
                            
                        if canClassBeAdded(coursesSplit[d][0], semesters, i):           #making sure can be added
                            semesters[i] = np.append(semesters[i], coursesSplit[d][0])  #add the current course to the plan
                            courseCount +=1                                             #inc total course count
                            coursesSplit[d] = np.delete(coursesSplit[d], 0)             #deleting from course dict after added
                            if courseCount % avgSemCourses == 0:                        #check if need to change semesters
                                i+=1
                        else:
                            lastClass = coursesSplit[d][0]
                              
#----------------------------------------------------------------------------------------------------------------------------                                
                elif i == 1:
                    if classAddedInNextSem:
                        courseCount += 1
                        classAddedInNextSem = False
                    prereqAdded = False
                    if hasPrereq(coursesSplit[d][0]):      #check if the individual course has prereqs
                        ph = getPrereqs(coursesSplit[d][0]) #if so we retrieve them
                        for pres in ph:                     #go through each prereq
                            if pres not in semesters[i] and pres not in semesters[i-1]:    #make sure prereq is not already in the course plan in this semester or the previous
                                semesters[i] = np.append(semesters[i], pres) #if not then append it to the course plan
                                coursesSplit = removeCourseFromDict(coursesSplit, pres) #after the courses is added we must remove it from course dictionary
                                prereqAdded = True                           #tell the program a prereq was aded
                                courseCount +=1                             #increase total course count by 1
                                if courseCount % avgSemCourses == 0:       #check if its need to go to next semester
                                    i+=1
                    
                    if not prereqAdded:                                                #if not prereq was added do the followiing
                        if hasCoreq(coursesSplit[d][0]):                            #check for coreqs
                            ch = getCoreqs(coursesSplit[d][0])                      #get coreqs
                            for cos in ch:                                         
                                if cos not in semesters[i] and cos not in semesters[i-1]:                         #if  coreq not in plan add it
                                    semesters[i] = np.append(semesters[i], cos)
                                    courseCount +=1                                 #inc total course count
                                    coursesSplit = removeCourseFromDict(coursesSplit, cos)
                                    if courseCount % avgSemCourses == 0:        
                                        i+=1
                        if coursesSplit[d][0] == lastClass:                                   #adding this to fix a infinite loop glitch that happens if only one department is left in the dictionary and the class trying to be added has a prereq in the same semester it is in so you must move to the next semester
                            semesters[i+1] = np.append(semesters[i+1], coursesSplit[d][0])
                            classAddedInNextSem = True
                            coursesSplit[d] = np.delete(coursesSplit[d], 0)
                            
                        if canClassBeAdded(coursesSplit[d][0], semesters, i):           #making sure can be added
                            semesters[i] = np.append(semesters[i], coursesSplit[d][0])  #add the current course to the plan
                            courseCount +=1                                             #inc total course count
                            coursesSplit[d] = np.delete(coursesSplit[d], 0)             #deleting from course dict after added
                            if courseCount % avgSemCourses == 0:                        #check if need to change semesters
                                i+=1
                        else:
                            lastClass = coursesSplit[d][0]

#-------------------------------------------------------------------------------------------------------------------------------                    
                elif i == 2:
                    if classAddedInNextSem:
                        courseCount += 1
                        classAddedInNextSem = False
                    prereqAdded = False
                    if hasPrereq(coursesSplit[d][0]):      #check if the individual course has prereqs
                        ph = getPrereqs(coursesSplit[d][0]) #if so we retrieve them
                        for pres in ph:                     #go through each prereq
#!!!!!!!!!!!THE IF STATEMENTS ARE GETTING FUTURE WARNING ERRORS, IF THEY BREAK CAN BE FIXED BY CONVERTING SEMESETER NP.ARR TO STRRING USING .astype(str)
                            if pres not in semesters[i].astype(str) and pres not in semesters[i-1].astype(str) and pres not in semesters[i-2].astype(str):    #make sure prereq is not already in the course plan in this semester or the previous
                                semesters[i] = np.append(semesters[i], pres) #if not then append it to the course plan
                                coursesSplit = removeCourseFromDict(coursesSplit, pres) #after the courses is added we must remove it from course dictionary
                                prereqAdded = True                           #tell the program a prereq was aded
                                courseCount +=1                             #increase total course count by 1
                                if courseCount % avgSemCourses == 0:       #check if its need to go to next semester
                                    i+=1
                    
                    if not prereqAdded:                                                #if not prereq was added do the followiing
                        if hasCoreq(coursesSplit[d][0]):                            #check for coreqs
                            ch = getCoreqs(coursesSplit[d][0])                      #get coreqs
                            for cos in ch:                                         
                                if cos not in semesters[i] and cos not in semesters[i-1] and cos not in semesters[i-2]:                         #if  coreq not in plan add it
                                    semesters[i] = np.append(semesters[i], cos)
                                    courseCount +=1                                 #inc total course count
                                    coursesSplit = removeCourseFromDict(coursesSplit, cos)
                                    if courseCount % avgSemCourses == 0:        
                                        i+=1 
                        if coursesSplit[d][0] == lastClass:                                   #adding this to fix a infinite loop glitch that happens if only one department is left in the dictionary and the class trying to be added has a prereq in the same semester it is in so you must move to the next semester
                            semesters[i+1] = np.append(semesters[i+1], coursesSplit[d][0])
                            classAddedInNextSem = True
                            coursesSplit[d] = np.delete(coursesSplit[d], 0)
                            
                        if canClassBeAdded(coursesSplit[d][0], semesters, i):           #making sure can be added
                            semesters[i] = np.append(semesters[i], coursesSplit[d][0])  #add the current course to the plan
                            courseCount +=1                                             #inc total course count
                            coursesSplit[d] = np.delete(coursesSplit[d], 0)             #deleting from course dict after added
                            if courseCount % avgSemCourses == 0:                        #check if need to change semesters
                                i+=1
                        else:
                            lastClass = coursesSplit[d][0]
                        
                          
 #------------------------------------------------------------------------------------------------------------------------------
                elif i == 3:
                    if classAddedInNextSem:
                        courseCount += 1
                        classAddedInNextSem = False
                    prereqAdded = False
                    if hasPrereq(coursesSplit[d][0]):      #check if the individual course has prereqs
                        ph = getPrereqs(coursesSplit[d][0]) #if so we retrieve them
                        for pres in ph:                     #go through each prereq
                            if pres not in semesters[i] and pres not in semesters[i-1] and pres not in semesters[i-2] and pres not in semesters[i-3]:    #make sure prereq is not already in the course plan in this semester or the previous
                                semesters[i] = np.append(semesters[i], pres) #if not then append it to the course plan
                                coursesSplit = removeCourseFromDict(coursesSplit, pres) #after the courses is added we must remove it from course dictionary
                                prereqAdded = True                           #tell the program a prereq was aded
                                courseCount +=1                             #increase total course count by 1
                                if courseCount % avgSemCourses == 0:       #check if its need to go to next semester
                                    i+=1
                    if not prereqAdded:                                                #if not prereq was added do the followiing
                        if hasCoreq(coursesSplit[d][0]):                            #check for coreqs
                            ch = getCoreqs(coursesSplit[d][0])                      #get coreqs
                            for cos in ch:
                                if cos not in semesters[i] and cos not in semesters[i-1] and cos not in semesters[i-2] and cos not in semesters[i-3]:                         #if  coreq not in plan add it
                                    semesters[i] = np.append(semesters[i], cos)
                                    courseCount +=1                                 #inc total course count
                                    coursesSplit = removeCourseFromDict(coursesSplit, cos)
                                    if courseCount % avgSemCourses == 0:        
                                        i+=1
                        
                        semesters[i] = np.append(semesters[i], coursesSplit[d][0])  #add the current course to the plan
                        courseCount +=1                                             #inc total course count
                        coursesSplit[d] = np.delete(coursesSplit[d], 0)             #deleting from course dict after added
                        if courseCount % avgSemCourses == 0:                        #check if need to change semesters
                            i+=1
                
                
                #semesters[i] = np.append(semesters[i], coursesSplit[d][0])     #retrieve the first value(course) and add it to the sem plan
                #courseCount +=1
                #coursesSplit[d] = np.delete(coursesSplit[d], 0)
                #if courseCount % avgSemCourses == 0:
                    #i+=1
                
            

    return semesters
   
#-------------------------------------------------------------- example
co = [ucla]
ma = ['Aerospace Engineering']
stR = getStudentRequiredCourses(co, ma)
print('Required Courses:')
print(stR)
print('\n\n')
n = createPlan(stR)
print('\n\nPlan:')
print('Sem 1')
print(n[0])
print('\nSem 2')
print(n[1])
print('\nSem 3')
print(n[2])
print('\nSem 4')
print(n[3])
```
